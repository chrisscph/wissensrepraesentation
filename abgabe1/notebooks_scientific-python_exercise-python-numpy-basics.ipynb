**Task:**

From each row, subtract the maximum value of that row.

```python
### Your Solution
z = np.random.randint(0,10, size=(3,4))
print(z)
for j in range(z.shape[0]):
    max_value = np.amax(z[j])
    for i in range(z.shape[1]):
        z[j][i] -= max_value   
print(z)
```

**Task:**
    
Divide each column by the sum of that column. Then verify that each column adds up to 1.

```python
### Your Solution
z = np.arange(12, dtype=float).reshape((3,4))
print(z)

sum_column = (z.sum(axis=0))

for j in range(z.shape[0]):
    for i in range(z.shape[1]):
        z[j][i] = z[j][i] / sum_column[i]
        
print(z)

sum_column = (z.sum(axis=0))
print(np.all(sum_column == 1))
```

**Task:**

Negate all elements between 3 and 8 in place.

```python
### Your Solution
Z = np.arange(11)

for i in Z:
    if Z[i] > 3 and Z[i] < 8:
        Z[i] *= (-1)
print(Z)
```
